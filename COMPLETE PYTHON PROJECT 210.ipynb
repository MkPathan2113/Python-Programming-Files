```python
# PROJECT OF DATA ANALYSIS USING PYTHON :

# NUMPY , PANDAS , MATPLOTLIB AND SEABORN LIBRARIES ARE USED IN THIS PROJECT ....

# WITH THE HELP OF THIS PROJECT WE CAN ANALYSE ANY TYPE OF DATA ......

# IN FUNCTION DEF WE IMPORTED ALL 4 LIBRARIES AS FOLLOWS...


import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
  
# WE ARE TAKING INPUT OF DATA FROM USER BY USING THE PATH OF THE FILE ......

# AFTER WE OUT THE READ CONDITION READ THAT IMPORTED DATA......

# NOW WE ARE GOING TO PUT CONDITIONS ONE BY ONE IN THE FORM OF NUMBER CHOICES......

# SO THAT USER CAN HANDLE THIS PROJECT EASILY BY CHOOSING ANY NUMBER.....

path=input("\n\nENTER THE PATH OF THE FILE :- \n\n\n")
df=pd.read_csv(path)
print(df)
    
    
# WE PUT 10 CHOICES FOR USER TO ANALYSE THE DATA EASILY ....

# HERE ARE THE CHOICES FOR THE USER .... 

# PRINT FUNCTION PRINTS THE LIST OF THE CONDITIONS SO THAT USER CAN PICK WHICH THEY WANT TO ANALYSE THERE DATASET ....

def PY_PROJECT():
    
    print("\n\n\n***_____PLEASE SELECT YOUR CHOICE_____***\n\n")

    print("PRESS-1. FOR STATISTICAL SUMMARY :- \n")
    
    print("PRESS-2. For UPPER FIVE ROWS :- \n")
    
    print("PRESS-3. FOR LOWER 5 ROWS :- \n")
    
    print("PRESS-4. TO FIND NULL VALUES :- \n")
    
    print("PRESS-5. TO FIND NULL VALUES IN EACH COLUMN SEPARETALY :- \n")
    
    print("PRESS-6. TO FIND THE MAXIMUM VALUE IN EVERY REGRESSIVE COLUMN :- \n")
    
    print("PRESS-7. FOR INFORMATION OF DATASET :- \n")
    
    print("PRESS-8. TO FIND THE MINIMUM VALUE IN EVERY REGRESSIVE COLUMN :- \n")
    
    print("PRESS-9. TO FIND LIST OF TYPES OF ALL COLUMNS :- \n")
    
    print("PRESS-10. TO DISPLAY RANDOMLY ANY NUMBER OF RECORDS IN DATASET :- \n")
    
    print("PRESS-11. EXIT :- \n")
    
    while True:
        choice=input("\n\n ENTER YOUR CHOICE:  \n\n YOUR CHOICE IS :- ")
    
    
    # THIS DF.DESCRIBE STATISTICAL SUMMARY FUNCTION GIVES THE MEAN , STD AND IQR VALUES...
    # FUNCTION EXCLUDES THE CHARACTER COLUMNS AND GIVEN SUMMARY ABOUT REGRESSIVE COLUMNS...

        if choice=='1':

            print(" \n THE STATICAL SUMMARY OF DATA : \n\n")

            print(df.describe())

    # THIS DF.HEAD FUNCTION IS USED TO FIND UPPER OR FIRST FIVE ROWS OF DATASET ....
    # IF USER WANTS TO SEE ONLY HEAD FIVE ROWS THEN HE CAN CHOOSE 2 ....

        elif choice=='2':

            print(" \n THE UPPER FIVE ROWS : \n\n")

            print(df.head())

    # THIS DF.TAIL FUNCTION IS USED TO FIND LOWER OR BOTTOM FIVE ROWS OF DATASET ....
    # IF USER WANTS TO SEE ONLY BOTTOM FIVE ROWS THEN HE CAN CHOOSE 3 ....

        elif choice=='3':

            print(" \n THE LOWER FIVE ROWS : \n\n")

            print(df.tail())

    # THIS ISNULL FUNCTION IS USED FOR FINDING THE NULL VALUES OF DATASET....
    # IF USER WANTS TO SEE ALL EMPTY VALUUES THEN HE CAN CHOOSE 4 ....

        elif choice=='4':

            print(" \n NULL VALUES IN DATASET : \n\n")

            print(df.isnull())

    # THIS FUNCTIN IS USED TO FIND THE SUM OF THE NULLL VALUES OR TOTAL NULL VALUES...        
    # THIS FUNCTION GIVES THE RESULT AS PER COLUMNS IN WHICH IT ANALYSE THAT IN EVARY COLUMN HOW MANY NULL VALUES ARE THERE ....

        elif choice=='5':

            print(" \n THE SUM OF NULL VALUES : \n\n")

            print(df.isnull().sum())

    # THIS FUNCTION GIVES THE MAXIMUM VALUES IN DATASET COLUMNWISE ....
    # SO THAT WE CAN EASILY GET THE BIGGEST VALUE IN EVERY COLUMN ....

        elif choice=='6':

            print(" \n THE MAX VALUE IN THE DATASET : \n\n")

            print(df.max())

    # THIS FUNCTION GIVES THE INFORMATION ABOUT THE DATAFRAME ....
    # INCLUDING THE INDEX DTYPE AND COLUMNS , NON-NULL VALUES AND MEMORY USAGE ....

        elif choice=='7':

            print(" \n INFORMATION OF DATASET : \n\n")

            print(df.info())

    # THIS FUNCTION GIVES THE MINIMUM VALUES IN DATASET COLUMNWISE ....
    # SO THAT WE CAN EASILY GET THE SMALLEST VALUE IN EVERY COLUMN ....

        elif choice=='8':

            print(" \n THE MIN VALUE IN THE DATASET : \n\n")

            print(df.min())

    # THIS FUNCTION RETURNS A SERIES WITH THE DATA TYPE OF EACH COLUMN ....
    # THE RESULT'S INDEX IS THE ORIGINAL DATAFRAME'S COLUMNS ....

        elif choice=='9':

            print(" \n THE DATA TYPES IN EACH COLUMN : \n\n")

            print(df.dtypes)

    # GIVES A RANDOM SAMPLE OF ITEMS FROM AN AXIS OF OBJECT ....
    # YOU CAN ALSO USE RANDOM_STATE REPRODUCIBILITY ....

        elif choice=='10':

            print(" \n RANDOM SAMPLE OF ITEM IN DATASET : \n\n")

            print(df.sample()) 
            
    # IF USER WANTS TO END UP THE CHOICES AND WANTS TO EXIT WITH HIS OUTPUT THEN HE CAN PRESS 11 ....

        elif choice=='11':
            print("\n\n_____YOU ARE EXIT NOW_____\n  _____ THANK YOU_____")
            break
            
    # IN THIS PROJECT WE IMPORTED 11 CONDITIONS ONE BY ONE ....
    
    # IF USER PUTS A CHOICE OUT OF THESE 11 THEN IT WILL PRINT THE BELOW STRING ....
    
        else:
            print('\n\n YOUR CHOICE IS INCORRECT. \n\n PLEASE ENTER CORRECT CHOICE...')

    
PY_PROJECT()
    
    
    # THANK YOU SO MUCH .....
        
          ####_____________ END _____________####
```
